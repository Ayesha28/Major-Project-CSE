# Calculating number of direct string match and printing all those names which does not have direct string match

```python
for i, row in Attendee_list.iterrows():
    Attendance_count+=1
    Attendee_name=f"{row['Attendee']}"
    upper_case_attendee_name=Attendee_name.upper()
    sorted_attendee_name= ''.join(sorted(upper_case_attendee_name))
    ans=900
    flag=0
    for j, row2 in Master_Student_list.iterrows():
        Master_name=f"{row2['Full Name']}"
        if(Master_name==upper_case_attendee_name):
            flag=1
            break
           
    if(flag==1):
        direct_list.append(Master_name)
        direct_string_match+=1
    else:
        levensht_list.append(Attendee_name)
        count_levenshtein_match+=1
        
levensht_list
```

```python
direct_list
```

```python
levenshtein_threshold_0=0
levenshtein_threshold_1=0
levenshtein_remaining=0
```

```python
for attend in levensht_list:
    attendee_name=attend
    ans=500
    upper_case_attendee_name=attendee_name.upper()
    sorted_attendee_name= ''.join(sorted(upper_case_attendee_name))
    for j, row2 in Master_Student_list.iterrows():
        Master_name=f"{row2['Full Name']}"
        sorted_master_name= ''.join(sorted(Master_name))
        dis=(enchant.utils.levenshtein(str(sorted_attendee_name), str(sorted_master_name)))
        if(dis<=ans):
            ans=dis
            resultant_master_name=Master_name
    print(attendee_name+"   "+resultant_master_name+"  "+str(ans))
```

# Calculating match for remaining string and marking levenshtein distance 0 and 1 as success and remaining as failure(string are first sorted and in between spaces are removed)

```python
for attend in levensht_list:
    attendee_name=attend
    ans=500
    upper_case_attendee_name=attendee_name.upper()
    upper_case_removed=upper_case_attendee_name.replace(" ", "")
    sorted_attendee_name= ''.join(sorted(upper_case_removed))
    for j, row2 in Master_Student_list.iterrows():
        Master_name=f"{row2['Full Name']}"
        master_name_removed=Master_name.replace(" ", "")
        sorted_master_name= ''.join(sorted(master_name_removed))
        dis=(enchant.utils.levenshtein(str(sorted_attendee_name), str(sorted_master_name)))
        if(dis<=ans):
            ans=dis
            resultant_master_name=sorted_master_name
            res_master=Master_name
            res_attendee=attend
    print(sorted_attendee_name+"   "+resultant_master_name+" "+res_master+" "+res_attendee+"  "+str(ans))
    if(ans==0):
        leven_success.append(res_master)
        levenshtein_threshold_0+=1
    elif(ans==1):
        leven_success.append(res_master)
        levenshtein_threshold_1+=1
    else:
        leven_failure.append(res_attendee)
        levenshtein_remaining+=1
```
